```python
import os
from glob import glob
import pandas as pd
from functools import reduce
from xml.etree import ElementTree as et
from shutil import move
```

```python
# load the data
x_list=glob(r'F:/project/image/*.xml')

# replace \\ 
a1=lambda x:x.replace('\\','/')
x_list=list(map(a1,x_list))

print(x_list)
```

```python
# reading the xml
# filename ,size(width,height),object(name,xmin,xmax,ymin,ymax)
# example

t1=et.parse('F:/project/image/02.xml')
r1=t1.getroot()

# file name or img name
img=r1.find("filename").text

#width&height
w1=r1.find("size").find('width').text
h1=r1.find('size').find('height').text

#inside the list
flist=[]

# objects
o1=r1.findall('object')
for o2 in o1:
    name=o2.find('name').text
    bndbox=o2.find('bndbox')
    xmin=bndbox.find('xmin').text
    ymin=bndbox.find('ymin').text
    xmax=bndbox.find('xmax').text
    ymax=bndbox.find('ymax').text
    flist.append([img,w1,h1,name,xmin,ymin,xmax,ymax])

print(flist)
```

```python
# function
def ext_info(fname):
    t1=et.parse(fname)
    r1=t1.getroot()
    img=r1.find("filename").text
    w1=r1.find("size").find('width').text
    h1=r1.find('size').find('height').text
    flist=[]
    o1=r1.findall('object')
    for o2 in o1:
        name=o2.find('name').text
        bndbox=o2.find('bndbox')
        xmin=bndbox.find('xmin').text
        ymin=bndbox.find('ymin').text
        xmax=bndbox.find('xmax').text
        ymax=bndbox.find('ymax').text
        flist.append([img,w1,h1,name,xmin,ymin,xmax,ymax])
    
    return flist
```

```python
all_info=list(map(ext_info,x_list))
```

```python
all_info
```

```python
# 3d to 2d form
data=reduce(lambda x,y:x+y,all_info)
data
```

```python
# convert to dataframe
df=pd.DataFrame(data,columns=['filename','width','height','name','xmin','ymin','xmax','ymax'])
```

```python
#type conversion
c1=['width','height','xmin','ymin','xmax','ymax']
df[c1]=df[c1].astype(int)
```

```python
# conversation to yolo input
df['center_x']=((df['xmax']+df['xmin'])/2)/df['width'] #center_x
df['center_y']=((df['ymax']+df['ymin'])/2)/df['height'] #center_y
df['w']=(df['xmax']-df['xmin'])/df['width'] #weidth
df['h']=(df['ymax']-df['ymin'])/df['height'] #height
```

```python
# adding the id name 
def label(x):
    l1={'person':0,'hard hat':1,'high visibility vest':2,'no mask':3,'no hard hat':4,'mask':5,'no high Visibility vest':6,'id card':7}
    return l1[x]
```

```python
train['id']=train['name'].apply(label)
test['id']=test['name'].apply(label)
```

```python
#creating a folder
trf1=r'F:\project\image\train'
tef2=r'F:\project\image\test'

os.mkdir(trf1)
os.mkdir(tef2)
```

```python
c3=['filename','id','center_x','center_y','w','h']
train2=train[c3].groupby('filename')
test2=test[c3].groupby('filename')
```

```python
#save the image/value in folder
def save(fname,fpath,group):
    src=os.path.join(r'F:\project\image',fname)
    dest=os.path.join(fpath,fname)
    move(src,dest)

    tfile=os.path.join(fpath,os.path.splitext(fname)[0]+'.txt')
    group.get_group(fname).set_index('filename').to_csv(tfile,sep=' ',index=False,header=False)
```

```python
# convert to series
fs1=pd.Series(train2.groups.keys())
```

```python
fs1.apply(save,args=(trf1,train2,))
```
